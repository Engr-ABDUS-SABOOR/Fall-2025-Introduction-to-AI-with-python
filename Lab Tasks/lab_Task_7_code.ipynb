# 1)  Smart Home Energy Meter:-
A smart home meter logs real-time energy usage  
{    "time": "18:32",  
    "grid_power": 720,  
    "solar_power": 340,  
    "battery_soc": 67,  
    "appliance_status": {"AC": "OFF", "Fridge": "ON", "Motor": "OFF"}}  
1.	Print the "grid_power" value.  
2.	Add a new appliance "TV": "ON" inside "appliance_status".  
3.	Change "battery_soc" to 72.  
4.	Check if "solar_power" exists in the dictionary.  
5.	Delete the "Motor" key from "appliance_status".  


```python
# Smart Home Energy Meter Dictionary
meter = {
    "time": "18:32",
    "grid_power": 720,
    "solar_power": 340,
    "battery_soc": 67,
    "appliance_status": {"AC": "OFF", "Fridge": "ON", "Motor": "OFF"}
}

```

```python
#  grid_power value
print("Grid Power:", meter["grid_power"])


```

```python
# 2. Add a new appliance TV: ON
meter["appliance_status"]["TV"] = "ON"
print(meter)
```

```python
# 3. Change battery_soc to 72
meter["battery_soc"] = 72
print(meter)
```

```python
# 4. Check if solar_power exists in dictionary
print("solar_power exists:", "solar_power" in meter)

```

```python
# 5. Delete the Motor key
del meter["appliance_status"]["Motor"]
print(meter["appliance_status"])
```

```python
# Print final updated dictionary
print("Final Updated Meter Dictionary:\n",meter)
```

 # 3)  Weather Monitoring Station  
Hourly weather data is stored as a dictionary:  
{   "hour": 14,  
    "temperature": 33.8,  
    "humidity": 52,  
    "wind": {"speed": 12, "direction": "NE"},  
    "alerts": []}  

1.Print the wind direction.  
2.Add a new alert "High Temperature".  
3.Change the wind speed to 15.  
4.Use .keys() to list all top-level keys.  
Remove "humidity" using .pop().  

```python
weather = {
    "hour": 14,
    "temperature": 33.8,
    "humidity": 52,
    "wind": {"speed": 12, "direction": "NE"},
    "alerts": []
}

```

```python
# 1. Print the wind direction
print("Wind Direction:", weather["wind"]["direction"])


```

```python
# 2. Add a new alert "High Temperature"
weather["alerts"].append("High Temperature")

```

```python
# 3. Change the wind speed to 15
weather["wind"]["speed"] = 15

```

```python
# 4. List all top-level keys using .keys()
print("Top-level keys:", list(weather.keys()))

```

```python
# 5. Remove "humidity" using .pop()
weather.pop("humidity")

```

```python
# Print final updated weather dictionary
print("Final Updated Weather Data:")
print(weather)
```

# 4  Inventory System for Electronics Shop  
A shop tracks an item in its inventory:  
{  
    "item_id": 1042,  
    "name": "PWM Solar Charge Controller",  
    "stock": 18,  
    "price": 2500,  
    "features": {"voltage": "12/24V", "current": "20A", "brand": "SolarTech"}  
}  
1.	Print the brand of the controller.  
2.	Reduce the stock by 2 and update the dictionary.  
3.	Add a new feature "warranty": "1 year".  
4.	Check if "price" exists in the dictionary.  
5.	Merge another dictionary {"discount": 10} into the inventory item.  


```python
# Inventory item dictionary
inventory = {
    "item_id": 1042,
    "name": "PWM Solar Charge Controller",
    "stock": 18,
    "price": 2500,
    "features": {
        "voltage": "12/24V",
        "current": "20A",
        "brand": "SolarTech"
    }
}
```

```python
# 1. Print the brand of the controller
print("Brand:", inventory["features"]["brand"])

```

```python
# 2. Reduce the stock by 2
inventory["stock"] = inventory["stock"] - 2
print("Updated Stock:", inventory["stock"])
```

```python
# 3. Add a new feature "warranty": "1 year"
inventory["features"]["warranty"] = "1 year"
```

```python

# 4. Check if "price" exists in the dictionary
if "price" in inventory:
    print("Price exists in the inventory.")
else:
    print("Price does not exist.")
```

```python
# 5. Merge another dictionary into inventory
extra_info = {"discount": 10}
inventory.update(extra_info)
```

```python
# Print final inventory
print("Final Inventory Data:")
print(inventory)
```

# 5)  Hospital Patient Monitoring System  
Patient vitals are recorded every 10 minutes:  
{  
    "id": 907,  
    "name": "Faisal",  
    "heart_rate": 82,  
    "oxygen_level": 97,  
    "medications": {"aspirin": "8AM", "vitamin_D": "12PM"}  
}  
1)	Print the oxygen level.  
2)	Add a new medication "antibiotic": "6PM".  
3)	Update heart rate to 78.  
4)	Get the medication schedule for "insulin" using .get() with default "Not Prescribed".  
5)	Use .items() to print all key-value pairs.  




























```python
# Patient vitals dictionary
patient = {
    "id": 907,
    "name": "Faisal",
    "heart_rate": 82,
    "oxygen_level": 97,
    "medications": {
        "aspirin": "8AM",
        "vitamin_D": "12PM"
    }
}
```

```python

# 1. Print the oxygen level
print("Oxygen Level:", patient["oxygen_level"])
```

```python
# 2. Add a new medication "antibiotic": "6PM"
patient["medications"]["antibiotic"] = "6PM"
```

```python
# 3. Update heart rate to 78
patient["heart_rate"] = 78
```

```python
# 4. Get medication schedule for "insulin" using .get()
insulin_time = patient["medications"].get("insulin", "Not Prescribed")
print("Insulin Schedule:", insulin_time)
```

```python
# 5. Print all key-value pairs using .items()
print("\nPatient Record:")
for key, value in patient.items():
    print(key, ":", value)
```
