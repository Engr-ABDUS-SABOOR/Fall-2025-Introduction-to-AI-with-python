Load Demand Data Recording  
An electrical feeder’s hourly load demand (kW) needs to be recorded for monitoring.  
# Tasks: 1
	Create a CSV file to store hourly load data  
	Write load values with time stamps  
	Organize each record on a separate line  


```python
import csv

with open("load_demand.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Time", "Load_kW"])
    writer.writerow(["08:00", 120.5])
    writer.writerow(["09:00", 135.2])
    writer.writerow(["10:00", 142.8])
    writer.writerow(["11:00", 150.4])
    writer.writerow(["12:00", 158.9])
    print("Load demand data saved")
```

Electricity Billing Record Creation  
A housing society wants to store monthly electricity billing data.  
# Tasks: 5
	Create a billing data file  
	Write customer billing records  
	Append new monthly data without overwriting existing records  


```python
import csv
with open("electricity_billing.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Customer_ID", "Month", "Units", "Amount"])
    writer.writerow(["C001", "Jan", 250, 1875])
    writer.writerow(["C002", "Feb", 275, 2060])
    writer.writerow(["C003", "Mar", 260, 1950])
    writer.writerow(["C004", "Apr", 290, 2175])
    writer.writerow(["C005", "May", 310, 2325])
    print("Electricity billing data saved")
```

# Task: 7
Organize and Process CSV Data  
Read the CSV file line by line and:   
	Count the total number of recorded days    
	Display each record with a serial number  
The header must not be counted as a data record.  


```python
import csv
with open("solar_generation.csv", "r") as src:
    reader = csv.reader(src)
    header = next(reader)
    
    print("S.No | Date | Energy_kWh")
    count = 0
    for i, row in enumerate(reader, start=1):
        print(f"{i} | {row[0]} | {row[1]}")
        count += 1
        print("\nTotal recorded days:", count)
```
