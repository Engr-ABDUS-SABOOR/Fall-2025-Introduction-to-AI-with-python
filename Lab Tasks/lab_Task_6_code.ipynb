 Access each element by index

## (Task 4):
 Nested Tuples & Mixing

 Access the date tuple and print year, month, day

```python

date = data[1]  # nested tuple at index 1
year = date[0]
month = date[1]
day = date[2]
print("Year:", year)
print("Month:", month)
print("Day:", day)

```
