# Task 1:
A lab keeps records of resistor values.Students need to print each resistance value for verification.  
Print each resistance value from a list of resistors.  
Create a list, then use for loop to display each item.  

```python
# Task 1: Print each resistance value

# Step 1: Create a list of resistors (in ohms)
resistors = [100, 220, 330, 470, 560]

# Step 2: Use a for loop to print each resistor
print("Resistor Values (Ohms):")
for a in resistors:
    print(a)
```

# Task 2:
Current readings are stored in a list. Engineers want to calculate the square of each value for power loss calculation.  
Print the square of each current reading using formula I².   
Make a list of current values, loop and multiply each by itself.   

```python
# Task 2: Calculate square of current readings

# Step 1: Create a list of current readings (in Amperes)
currents = [2, 3, 4, 5, 6]

# Step 2: Print the square of each current using I^2
for i in currents:
    power_loss = i * i   # I^2 formula
    print(f"Current: →  {power_loss} ")
```

Print double of each capacitor value from a list.   

```python
# Task 5: Print double of each capacitor value

# Step 1: Create a list of capacitor values (in microfarads)
capacitors = [10, 22, 33, 47, 56]

# Step 2: Use a for loop to print double of each capacitor
for c in capacitors:
    double_value = c * 2
    print(f"Original: {c} μF → Doubled: {double_value} μF")
```

# Task 7:
             A solar system records daily energy in kWh. Students must calculate and print half of each value (to simulate half-load usage).   
Print half of each solar energy reading from a list.  

```python
# Task 7: Print half of each solar energy reading

# Step 1: Create a list of daily solar energy readings (in kWh)
solar_energy = [12, 15, 20, 18, 25]

# Step 2: Use a for loop to calculate and print half of each value
for energy in solar_energy:
    half_energy = energy / 2
    print(f"Original: {energy} kWh → Half-load: {half_energy} kWh")
```

# Task 1:
During charging, the battery slowly increases its percentage. Students want to simulate this charging process in code.  
Keep printing "Charging..." until battery percentage reaches 100%.  
Start with battery = 0, increase by 10 in each loop, stop when 100.  

```python
# Task 1: Simulate battery charging

# Step 1: Start battery at 0%
battery = 0

# Step 2: Use a while loop to increase battery until 100%
while battery < 100:
    battery += 10
    print(f"Charging... Battery at {battery}%")

print("Battery fully charged!")
```

# Task 2:
Engineers use countdown timers for testing. Students must simulate a timer that runs before starting a test.  
Countdown timer from 10 to 0, showing "Test Running".  
Start variable = 10, decrease each loop, stop at 0.  

```python
# Task 2: Countdown timer simulation

# Step 1: Start the countdown at 10
timer = 10

# Step 2: Use a while loop to count down to 0
while timer >= 0:
    print(f"Timer: {timer} → Test Running")
    timer -= 1  # Decrease by 1 each loop

print("countdown stop")
```

 # Task 3:
A voltage monitoring system asks for user input repeatedly until the system is turned off.  
Keep asking user for voltage until they enter 0.  
Use input() in while loop, stop when entered value is 0.  

```python
# Step 1: Initialize voltage
voltage = None

# Step 2: Keep asking user until they enter 
while voltage != 0:
    voltage = float(input("Enter voltage reading (V): "))
    
    if voltage != 0:
        print(f"Voltage recorded: {voltage} V")
    else:
        print("System turned off.")
```

 # Task 4:
Power in a system doubles when load increases. Students need to simulate this doubling until it reaches a limit.  
Keep doubling the power until it exceeds 1000 W.  
Start power = 1, multiply by 2 each loop, stop when >1000.  

```python
# Task 4: Simulate power doubling

# Step 1: Start power at 1 W
power = 1

# Step 2: Use a while loop to double power until it exceeds 1000 W
while power <= 1000:
    print(f"Current Power: {power} W")
    power *= 2  # Double the power each loop

print("Power limit exceeded!")
```

# Task 5:
            A machine keeps running until its cycle count reaches a limit. Students must simulate this with a loop.  
Start from cycle = 1 and keep printing "Cycle Running" until cycle reaches 5.  

```python
# Task 5: Simulate machine cycles

# Step 1: Start from cycle 1
cycle = 1

# Step 2: Use while loop to run cycles until 5
while cycle <= 5:
    print(f"Cycle {cycle} Running")
    cycle += 1  # Increase cycle count

print("All cycles completed!")
```

# Task 6:
            A motor runs until its speed reaches 3000 RPM. Students must simulate this by starting from 0 and adding 500 each loop.  
Keep increasing speed by 500 until it reaches 3000, printing the speed each time.  

```python
# Task 6: Simulate motor speed increase

# Step 1: Start motor speed at 0 RPM
speed = 0

# Step 2: Use while loop to increase speed by 500 until 3000 RPM
while speed < 3000:
    speed += 500
    print(f"Motor Speed: {speed} RPM")

print("Motor reached target speed!")
```
