# 4. Task: 
Bit Parity Checker   
Digital systems often use parity bits to detect errors in data transmission.   
‚Ä¢ Even Parity ‚Üí If the number of 1s in a binary signal is even.   
‚Ä¢ Odd Parity ‚Üí If the number of 1s in a binary signal is odd.   
## Task Description   
Takes a binary signal stored as an integer (e.g., 0b11010110).   
Counts the number of 1s in the signal using the .bit_count() method.   
Determines whether the signal has even parity or odd parity.   
Prints the result clearly.   

```python
signal = 0b11010110
ones_count = signal.bit_count()
parity = "Even" if ones_count % 2 == 0 else "Odd"
print("Binary Signal:", bin(signal))
print("Number of 1s:", ones_count)
print("Parity:", parity)
```

Note:   
import cmath

## Circuit parameters   
R = 50        # ohms

C = 100e-6    # F

w = 2 * math.pi * f   
 
## Impedances

```python
import cmath
import math
# Circuit parameters
R = 50        # ohms
L = 0.1       # henry
C = 100e-6    # farad
f = 60        # hertz
```

```python
# Angular frequency
w = 2 * math.pi * f
# Impedances
Z_R = complex(R, 0)             # Resistor impedance
Z_L = complex(0, w * L)         # Inductor impedance (jœâL)
Z_C = complex(0, -1 / (w * C))  # Capacitor impedance (-j/œâC)
```

```python
# Print results
print("Resistor Impedance (Z_R):", Z_R)
print("Inductor Impedance (Z_L):", Z_L)
print("Capacitor Impedance (Z_C):", Z_C)
```

# 6. Task 
Find Total Impedance    
Add the three impedances.   
Print real part (resistive) and imaginary part (reactive).   
Print magnitude and phase angle (use abs() and cmath.phase())  

```python

import cmath
Z1 = 7 + 5j
Z2 = 9 - 3j
Z3 = 6 + 5j
```

```python
# i. Add the three impedances
Z_total = Z1 + Z2 + Z3
print("Total Impedance:", Z_total)
print("Resistive Part (Real):", Z_total.real)
print("Reactive Part (Imag):", Z_total.imag)
```

```python
# iii. Magnitude and phase angle
magnitude = abs(Z_total) 
phase_angle = cmath.phase(Z_total)      
print("Magnitude:", magnitude)
print("Phase Angle (radians):", phase_angle)
print("Phase Angle (degrees):", phase_angle * 180 / 3.14159)
```
