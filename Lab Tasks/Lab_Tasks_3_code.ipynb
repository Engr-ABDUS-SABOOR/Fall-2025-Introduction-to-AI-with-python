# 1.	Task: Voltage Level Classification   
i.	A text file contains circuit voltage readings (one per line).  
ii.	Task: Read the file and classify each voltage as:  
a.	Low (< 110V)  
b.	Medium (110V–220V)  
c.	High (> 220V)  
iii.	Save the classification results into a new file.  


```python
with open('voltages.txt', 'r') as infile:
    voltage_readings = infile.readlines()


classification_results = []


for line in voltage_readings:
    try:
        voltage = float(line.strip())  

        
        if voltage < 110:
            classification = "Low"
        elif 110 <= voltage <= 220:
            classification = "Medium"
        else:
            classification = "High"

        
        classification_results.append(f"{voltage}V: {classification}")

    except ValueError:
       
        classification_results.append(f"Invalid reading: {line.strip()}")

with open('classified_Voltages.txt', 'w') as outfile:
    for result in classification_results:
        outfile.write(result + '\n')

print("Voltage classification complete. Results saved to 'classified_voltages.txt'.")
```

# 2.	Task: Pass/Fail Test Report (Engineering Labs)  
i.	A file stores students’ lab test scores.  
ii.	Task: Read scores from file and apply conditions:  
a.	Pass if ≥ 50.  
b.	Fail if < 50.  


```python
with open(r'score.txt', 'r') as infile:
    score_lines = infile.readlines()

results = []


for line in score_lines:
    try:
        score = float(line.strip()) 

       
        if score >= 50:
            status = "Pass"
        else:
            status = "Fail"

        
        results.append(f"{score}: {status}")

    except ValueError:
        
        results.append(f"Invalid score: {line.strip()}")


with open(r'test_report.txt', 'w') as outfile:
    for result in results:
        outfile.write(result + '\n')

print("Test report generated successfully in 'test_report.txt'.")
```

# 3.	Task: Current Rating Check (Electrical Machines)
i.	A file has current values (in Amperes).  
ii.	Task: Check for each line:  
a.	If current < 5A → Safe  
b.	If 5A ≤ current < 10A → Warning  
c.	If ≥ 10A → Overload  


```python
# Open input file (with currents)
with open("current.txt", "r") as infile:
    for line in infile:
        # Remove spaces, newline, and commas
        line = line.strip().replace(",", "")

        # Convert text to number
        current = float(line)

        # Checking condition
        if current < 5:
            result = "Safe"
        elif 5 <= current < 10:
            result = "Warning"
        else:
            result = "Overload"

        # Results
        print(f"{current} - {result}")
```

# 4.	Task: File Extension Validator (General Engineering Reports)
i.	A directory has multiple engineering reports.  
ii.	Task: Write a program to scan all files.  
a.	If file ends with .pdf → Valid Report  
b.	If file ends with .docx → Needs Conversion  
c.	Otherwise → Invalid Format  
iii.	Count and display the numbers of each type.  


```python
import os

folder = "report"   # folder name 

pdf = docx = invalid = 0  # counters

for file in os.listdir(folder):
    if file.endswith(".pdf"):
        print(f"{file} - Valid Report")
        pdf += 1
    elif file.endswith(".docx"):
        print(f"{file} - Needs Conversion")
        docx += 1
    else:
        print(f"{file} - Invalid Format")
    invalid+=1
```

# 11.	Task:
Electrical systems run at a normal frequency of 50 or 60 Hz. Frequency outside this range may damage the equipment.  
Check frequency value (Hz):  
•	If < 40 → "Too Low"  
•	Elif between 40–60 → "Normal Range"  
•	Elif > 60 → "Too High"  
•	Store frequency, check conditions step by step.  


```python
# Store frequency (Hz) in a variable (decimals allowed)
frequency = float(input("Enter frequency in Hz: "))

# Check the frequency step by step
if frequency < 40:
    print("Too Low")
elif 40 <= frequency <= 60:   # 40 and 60 are considered within the normal range
    print("Normal Range")
else:                        # frequency > 60
    print("Too High")
```

Check lux value and print lighting condition:

•	Elif lux < 300 → "Dim Light – Lights ON low"
