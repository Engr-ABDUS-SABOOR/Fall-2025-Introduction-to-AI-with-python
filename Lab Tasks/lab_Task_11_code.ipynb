 # Task 1:
 Map Device Names to Their Power Ratings  
Given:  
devices = ["Fan", "Light", "AC", "Pump"]  
power = [60, 15, 1800, 750]  
Create a dict using comprehension:  
{"Fan": 60, "Light": 15, ...}  

```python
#  lists
devices = ["Fan", "Light", "AC", "Pump"]
power = [60, 15, 1800, 750]

```

```python
# Create a dictionary using comprehension
device_power_dict = {devices[i]: power[i] for i in range(len(devices))}


```

```python

print("Device Power Ratings:")
print(device_power_dict)
```

# Task 2: 
Categorize Resistances as “Low” or “High”  
Resistance values (ohms):  
R = [5, 33, 47, 82, 10]  
Create a dict:  
Key = value, Value = "Low" if R < 40 else "High".  

```python
# resistance values in ohms
R = [5, 33, 47, 82, 10]

```

```python
# Create dictionary with value as key, category as value
resistance_category = {r: "Low" if r < 40 else "High" for r in R}


```

```python
# Print the resulting dictionary
print("Resistance Categories:")
print(resistance_category)
```

# Task 3:
Compute Current for Each Load  
Loads in watts:  
loads = [60, 100, 150, 200]   
Assume voltage = 200 V  
Dictionary comprehension to produce:  
{load: load/200} (I = P/V)  

```python
#  load values in watts
loads = [60, 100, 150, 200]


```

```python
# voltage in volts
voltage = 200
```

```python

# current for each load (I = P / V) using dictionary comprehension
current_dict = {load: load / voltage for load in loads}

```

```python

print("Current for Each Load (I = P/V):")
print(current_dict)
```

# Task 4: Convert Dictionary Values From °C to °F  
Given:  
temps_c = {"motor": 40, "transformer": 60, "room": 28}  
Use dict comprehension to convert to Fahrenheit.  


```python
temps_c = {"motor": 40, "transformer": 60, "room": 28}

# Convert °C to °F using dictionary comprehension
temps_f = {k: (v * 9/5) + 32 for k, v in temps_c.items()}

print("Temperatures in °F:", temps_f)
```

# Task 5: Create Dict of Components With Index Keys  
Given component names:  
components = ["Resistor", "Capacitor", "Inductor", "Diode"]  
Create {0:"Resistor", 1:"Capacitor", ...}  
Using dict comprehension + enumerate.  


```python
components = ["Resistor", "Capacitor", "Inductor", "Diode"]

# Create dictionary with index as key
components_dict = {i: comp for i, comp in enumerate(components)}

print("Components with index keys:", components_dict)
```


# Task 6: Filter Only Devices With Power > 100W  
Given:  
device_power = {"Fan": 60, "Light": 15, "Heater": 1200, "Iron": 1000, "Router": 10}  
Use dict comprehension to keep only items where value > 100.  


```python
device_power = {"Fan": 60, "Light": 15, "Heater": 1200, "Iron": 1000, "Router": 10}

# Keep only devices with power > 100W
high_power_devices = {k: v for k, v in device_power.items() if v > 100}

print("Devices with power > 100W:", high_power_devices)
```
