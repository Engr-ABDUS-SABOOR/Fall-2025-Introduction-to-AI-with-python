# 3. Lambda With map()
Use map + lambda to transform a list of numbers.  
Use map + lambda to transform a list of strings.  
Use map + lambda on a list of tuples.  

```python
numbers = [1, 2, 3, 4, 5]

# multiply each number by 2
doubled = list(map(lambda x: x * 2, numbers))

print("Original Numbers:", numbers)
print("Doubled Numbers:", doubled)
names = ["ali", "umar", "sara"]

# convert each name to uppercase
upper_names = list(map(lambda s: s.upper(), names))

print("\nOriginal Names:", names)
print("Uppercase Names:", upper_names)
pairs = [(2, 3), (4, 5), (6, 7)]

# add both elements inside each tuple → (a, b) → a + b
sums = list(map(lambda t: t[0] + t[1], pairs))

print("\nOriginal Tuples:", pairs)
print("Sum of Each Tuple:", sums)


```

# 4. Lambda With filter()
Use filter + lambda to keep only even numbers.  
Use filter + lambda to keep only strings longer than 3 characters.  
Use filter + lambda on a mixed list of different types.  

```python
numbers = [1, 2, 3, 4, 5, 6, 7, 8]

even_numbers = list(filter(lambda x: x % 2 == 0, numbers))

print("Original Numbers:", numbers)
print("Even Numbers:", even_numbers)

# 2. filter + lambda to keep ONLY strings longer than 3 characters

names = ["ali", "umar", "zain", "sara", "to"]

long_names = list(filter(lambda s: len(s) > 3, names))

print("\nOriginal Names:", names)
print("Names Longer Than 3 Characters:", long_names)

mixed_list = [1, "hello", 3.5, 2, "abc", 10]

only_integers = list(filter(lambda x: isinstance(x, int), mixed_list))

print("\nOriginal Mixed List:", mixed_list)
print("Only Integers:", only_integers)

```

# 5. Lambda With sorted() key  
Sort a list by length using lambda as key.  
Sort a list of tuples by their second element.  
Sort dictionary items by their values using lambda.  

```python
words = ["apple", "kiwi", "banana", "cat"]

# Sort by length of each word
sorted_by_length = sorted(words, key=lambda x: len(x))

print("Original Words:", words)
print("Sorted by Length:", sorted_by_length)

# 2. Sort a list of TUPLES by their SECOND element


pairs = [(1, 5), (3, 2), (7, 1), (4, 9)]

# Sort by tuple's second value (index 1)
sorted_by_second = sorted(pairs, key=lambda t: t[1])

print("\nOriginal Tuples:", pairs)
print("Sorted by Second Element:", sorted_by_second)


# 3. Sort dictionary items by their VALUES using lambda


scores = {"ali": 30, "sara": 50, "umar": 40}

# Convert dict to list of tuples, then sort by value
sorted_scores = sorted(scores.items(), key=lambda item: item[1])

print("\nOriginal Dictionary:", scores)
print("Sorted by Values:", sorted_scores)

```

# 6. Lambda With reduce()  
Use reduce + lambda to combine all numbers in a list.  
Use reduce + lambda to join multiple strings.  
Use reduce + lambda to find a custom smallest or largest element.  

```python
from functools import reduce

print("---- 1. Reduce + Lambda : Sum of numbers ----")

numbers = [2, 4, 6, 8, 10]

# Reduce to sum all numbers
total_sum = reduce(lambda a, b: a + b, numbers)
print("Sum of numbers:", total_sum)



print("\n---- 2. Reduce + Lambda : Join strings ----")

words = ["AI", "is", "very", "powerful"]

# Reduce to join strings with space
joined_text = reduce(lambda a, b: a + " " + b, words)
print("Joined string:", joined_text)



print("\n---- 3. Reduce + Lambda : Find largest number ----")

nums = [12, 5, 27, 9, 32, 15]

# Reduce to find maximum
largest = reduce(lambda a, b: a if a > b else b, nums)
print("Largest number:", largest)
```

# 8. Replace a small def with a lambda and compare readability  
Given this function:  
def label_meter(id):  
    return "Meter-" + str(id).zfill(3)  
Rewrite it using a lambda.  
Call both versions with 12.  
Write one short sentence on which is clearer.  

```python
# --- Normal function version ---
def label_meter(id):
    return "Meter-" + str(id).zfill(3)

# --- Lambda version ---
label_meter_lambda = lambda id: "Meter-" + str(id).zfill(3)

# --- Calling both with 12 ---
print("Using normal function:", label_meter(12))
print("Using lambda:", label_meter_lambda(12))

# --- One sentence comparison ---
print("\nThe normal function (def) is clearer and more readable than the lambda version.")
```

# 9. Sort electrical components using lambda key  
Sort this list of components by their watt rating using sorted() and a lambda key:  
components = [("LED", 5), ("Fan", 60), ("Motor", 350), ("Bulb", 40)]  
Then show the same sorting using operator.itemgetter.  
State one advantage of itemgetter.  

```python
from operator import itemgetter

# Given components list
components = [("LED", 5), ("Fan", 60), ("Motor", 350), ("Bulb", 40)]

# --- Sorting using lambda key ---
sorted_by_watt_lambda = sorted(components, key=lambda item: item[1])

# --- Sorting using itemgetter ---
sorted_by_watt_itemgetter = sorted(components, key=itemgetter(1))

# --- Printing results ---
print("Sorted using lambda:")
print(sorted_by_watt_lambda)

print("\nSorted using itemgetter:")
print(sorted_by_watt_itemgetter)

# --- Advantage of itemgetter ---
print("\nAdvantage: itemgetter is faster and cleaner than using lambda.")
```

# 10. Convert sensor strings using map() + lambda  
Convert a list of sensor voltage readings (strings) into floats and add +0.1 calibration offset:  
readings = ["2.5", "5.0", "3.3"]  
Show also the equivalent list comprehension.  
Say which one looks cleaner.  

```python
# Given readings
readings = ["2.5", "5.0", "3.3"]

# --- Using map + lambda ---
converted_map = list(map(lambda x: float(x) + 0.1, readings))

# --- Using list comprehension ---
converted_list_comp = [float(x) + 0.1 for x in readings]

# Printing results
print("Using map + lambda:", converted_map)
print("Using list comprehension:", converted_list_comp)

# Which one is cleaner?
print("\nList comprehension looks cleaner and easier to read.")
```

# 11. Filter out low power values using filter() + lambda  
From this list, keep only values above 50 W:  
powers = [20, 75, 45, 120, 10]  
Show the filter() version and a list-comprehension version.  

```python
# Given power list
powers = [20, 75, 45, 120, 10]

# --- Using filter + lambda ---
high_powers_filter = list(filter(lambda x: x > 50, powers))

# --- Using list comprehension ---
high_powers_listcomp = [x for x in powers if x > 50]

# Printing results
print("Using filter + lambda:", high_powers_filter)
print("Using list comprehension:", high_powers_listcomp)

# Which one is cleaner?
print("\nList comprehension is cleaner and easier to read.")
```

# 12. Use reduce() to combine values (total impedance calculation)  
Using functools.reduce + lambda, compute the product of impedances in this list:  
Z = [2, 3, 4]  
Then show the simpler alternative using a loop or math.prod.  
Write one short line on which is easier to understand.  

```python
from functools import reduce
import math

# Given impedances
Z = [2, 3, 4]

# --- Using reduce + lambda to calculate product ---
total_impedance_reduce = reduce(lambda a, b: a * b, Z)

# --- Using math.prod (simpler alternative) ---
total_impedance_prod = math.prod(Z)

# Printing results
print("Total impedance using reduce + lambda:", total_impedance_reduce)
print("Total impedance using math.prod:", total_impedance_prod)

# Which one is easier to understand?
print("\nmath.prod is simpler and easier to understand than reduce + lambda.")
```

# 13. Conditional lambda for load classification 
Write a lambda that returns "High Load" if power > 100 W, else "Low/Normal Load".  
Test it for 150 W and 90 W.  

```python
# Conditional lambda for load classification
load_classification = lambda power: "High Load" if power > 100 else "Low/Normal Load"

# Test with 150 W and 90 W
test1 = load_classification(150)
test2 = load_classification(90)

# Printing results
print("Power = 150 W:", test1)
print("Power = 90 W:", test2)
```
