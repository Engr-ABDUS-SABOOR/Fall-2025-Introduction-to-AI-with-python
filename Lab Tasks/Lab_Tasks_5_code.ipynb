#  Lab Task (NO 2)  
                  Using zip() to combine two lists element by element

# i️
                  Combine current and voltage readings

```python

currents = [2, 3, 4, 5]
voltages = [220, 230, 240, 250]

print("Current and Voltage Pairs:")
for i, v in zip(currents, voltages):
    print("Current:", i, "A |   Voltage:", v, "V")

print()  # empty line for clarity


```

# ii️ 
      Combine component names with their ratings

```python

components = ["Resistor", "Capacitor", "Inductor"]
ratings = ["220Ω", "100µF", "10mH"]

print("Component Ratings:")
for name, rating in zip(components, ratings):
    print(name, "→", rating)
```

##  Lab Task   (No 3): 

Using map() with lambda to apply a function to every list item

# i️ 
  Convert Celsius to Fahrenheit

```python

celsius = [0, 25, 30, 40]
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))
print("Temperatures in Celsius:", celsius)
print("Converted to Fahrenheit:", fahrenheit)

print()  # empty line for clarity


```

# ii️ 
    Convert Voltages (V) to Kilovolts (kV)

```python

voltages = [220, 110, 440, 330]
kilovolts = list(map(lambda v: v / 1000, voltages))
print("Voltages in Volts:", voltages)
print("Converted to Kilovolts:", kilovolts)
```

##  Lab Task  (No 4):
 Using filter() to select values based on a condition

# i️
   Current readings in Amps

```python

currents = [5, 12, 8, 15, 3, 20]

# Use filter() to keep only readings greater than 10 A
high_currents = list(filter(lambda x: x > 10, currents))
print("Current readings greater than 10 A:", high_currents)


```

# ii️ 
     Capacitor bank values in µF

```python

capacitors = [10, 20, 50, 100]

# Use filter() to select capacitors above 50 µF
large_capacitors = list(filter(lambda x: x > 50, capacitors))
print("Capacitors above 50 µF:", large_capacitors) 
```

#  Lab Task (NO 18):
   Using divmod() to get quotient and remainder

# i️ 
   Optical fiber cables for internet setup

```python

total_cables = 87
duct_capacity = 10

# divmod gives (quotient, remainder)
ducts_filled, cables_left = divmod(total_cables, duct_capacity)
print("Completed ducts:", ducts_filled)
print("Cables in last duct:", cables_left)
```

# ii️ 
    LEDs for DLD project

```python

total_leds = 42
leds_per_group = 12

groups, extra_leds = divmod(total_leds, leds_per_group)
print("Complete groups:", groups)
print("Extra LEDs left:", extra_leds)
```

#  Lab Task  (SSNO 19): 
   Using pow() to raise numbers to a power

# i️
   Current values in Amperes

```python

currents = [2, 3, 4, 5]

# Calculate the square of each current (I²)
squared_currents = [pow(i, 2) for i in currents]

# Print results
print("Current values:", currents)
print("Squared (I²) values for power calculation:", squared_currents)
```
