```python
import csv
# 1. Load the CSV manually
days = []
hours = []
consumptions = []
with open(r"C:\Users\PMLS\Desktop\ABDUS_SABOOR\AI_Lab\Assignment 2.csv") as file:
    reader = csv.reader(file)
    next(reader)   # skip header

    for row in reader:
        day = int(row[0])
        hour = int(row[1])
        cons = int(row[2])

        days.append(day)
        hours.append(hour)
        consumptions.append(cons)
```

```python
# 2. Group consumption by hour
hour_groups = {}   # {hour: [values]}

for i in range(len(hours)):
    h = hours[i]
    c = consumptions[i]

    if h not in hour_groups:
        hour_groups[h] = []
    hour_groups[h].append(c)

# 2.1 Average consumption per hour
hour_avg = {}
for h in hour_groups:
    hour_avg[h] = sum(hour_groups[h]) / len(hour_groups[h])

# 2.2 Peak consumption per hour
hour_peak = {}
for h in hour_groups:
    hour_peak[h] = max(hour_groups[h])

```

```python
# 2.3 Print results
print("\n===== Hourly Consumption Analysis =====")
for h in sorted(hour_groups):
    print(f"Hour {h}: Values = {hour_groups[h]}")
    print(f" → Average = {hour_avg[h]:.2f}")
    print(f" → Peak = {hour_peak[h]}")
    print()

```

```python
# 3. Group consumption by day

day_groups = {}   # {day: [values]}

for i in range(len(days)):
    d = days[i]
    c = consumptions[i]

    if d not in day_groups:
        day_groups[d] = []
    day_groups[d].append(c)

# 3.1 Total consumption per day
day_total = {}
for d in day_groups:
    day_total[d] = sum(day_groups[d])

# 3.2 Daily average
day_avg = {}
for d in day_groups:
    day_avg[d] = sum(day_groups[d]) / len(day_groups[d])

```

```python
print("\n===== Daily Consumption Analysis =====")
for d in sorted(day_groups):
    print(f"Day {d}: Values = {day_groups[d]}")
    print(f" → Total = {day_total[d]}")
    print(f" → Average = {day_avg[d]:.2f}")
    print()
```
