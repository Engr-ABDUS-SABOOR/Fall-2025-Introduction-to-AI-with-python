```python
import csv

consumption = []

with open(r"C:\Users\PMLS\Desktop\ABDUS_SABOOR\AI_Lab\Assignment 1.csv") as file:
    reader = csv.reader(file)

    next(reader)   #  this line SKIPS the header row

    for row in reader:
        # skip the empty rows
        if len(row) < 2:
            continue
        
        value = float(row[1].strip())
        consumption.append(value)



```

```python

print("Consumption list :", consumption)

total_consumption = sum(consumption)
average_consumption = total_consumption / len(consumption)
maximum_consumption = max(consumption)
minimum_consumption = min(consumption)
```

```python
# convert to kWh they give in watt we divided on 1000 
kwh_list = [x / 1000 for x in consumption]

print("Total consumption :", total_consumption)
print("Average consumption :", average_consumption)
print("Max :", maximum_consumption)
print("Min :", minimum_consumption)

print("Consumption in kWh :", kwh_list)
print("Total kWh :", sum(kwh_list))
print("Average kWh :", sum(kwh_list)/len(kwh_list))
```
