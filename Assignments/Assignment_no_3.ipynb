1.Load both CSVs.  
2.Store loads and prices into lists.  
3.Write a function to compute.  
3.1.Hourly bill  
3.2.Total bill  




```python
import csv
# 1. Load the CSV manually
loads = []
prices = []
with open(r"C:\Users\PMLS\Desktop\ABDUS_SABOOR\AI_Lab\Assignment 3 load.csv") as file:
    reader = csv.reader(file)
    next(reader)   # skip header

    for row in reader:
        loads.append(int(row[1]))

with open(r"C:\Users\PMLS\Desktop\ABDUS_SABOOR\AI_Lab\Assignment 3 price.csv") as file:
    reade = csv.reader(file)
    next(reade)   # skip header

    for row in reade:
        prices.append(int(row[0])  )


def compute_bills(loads, prices):
    hourly_bills = []      # list to store each hour bill
    total_bill = 0         # store total bill

    for load, price in zip(loads, prices):
        # Convert Watts â†’ kW
        load_kW = load / 1000

        # Bill formula
        bill = load_kW * price

        hourly_bills.append(bill)
        total_bill += bill

    return hourly_bills, total_bill
hourly_bills, total_bill = compute_bills(loads, prices)
print("Hourly Bills (Rs):")
for i, bill in enumerate(hourly_bills, start=1):
    print(f"Hour {i}: Rs {bill}")

print("\nTotal Bill (Rs):", total_bill)
```
