Find Total Impedance  
i. Add the three impedances.  
ii. Print real part (resistive) and imaginary part (reactive).  
iii. Print magnitude and phase angle (use abs() and cmath.phase()).  

```python

import cmath  # we used for complex numbers

#  Impedances
Z1 = 7 + 5j
Z2 = 9 - 3j
Z3 = 6 + 5j
```

```python
# i. Add the three impedances
Z_total = Z1 + Z2 + Z3
```

```python
# ii. Print real (resistive) and imaginary (reactive) parts
print("Total Impedance:", Z_total)
print("Resistive Part (Real):", Z_total.real)
print("Reactive Part (Imag):", Z_total.imag)
```

```python
# iii. Magnitude and phase angle
magnitude = abs(Z_total)
phase_angle_rad = cmath.phase(Z_total)
phase_angle_deg = phase_angle_rad * 180 / 3.14159
```

```python
print("Magnitude:", magnitude)
print("Phase Angle (radians):", phase_angle_rad)
print("Phase Angle (degrees):", phase_angle_deg)
```
