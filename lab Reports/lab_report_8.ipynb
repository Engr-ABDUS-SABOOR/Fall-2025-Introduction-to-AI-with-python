```python
#  Adjusted Power Readings
# Example: multiply each power reading by a calibration factor
power_readings = [10, 15, 20, 25]
calibration_factor = 1.05
adjust_power = list(map(lambda x: x * calibration_factor, power_readings))

```

```python

print(" Adjusted Power Readings ")
print("Original:", power_readings)
print("Adjusted:", adjust_power)
```

```python
#  Filter Noisy Sensor Data
# Remove sensor readings below a threshold (e.g., noise)
sensor_data = [0.1, 2.5, 0.05, 3.0, 0.0, 4.2]
threshold = 0.5
filtered_data = list(filter(lambda x: x > threshold, sensor_data))
```

```python
print(" Filtered Sensor Data ")
print("Original:", sensor_data)
print("Filtered:", filtered_data)
```

```python
#  Sort Components by Rating

# components as tuples (name, rating)
components = [("Resistor", 5), ("Capacitor", 10), ("Inductor", 3)]
# Sort by rating (second element of tuple)
sorted_components = sorted(components, key=lambda x: x[1])
```

```python
print(" Sorted Components by Rating ")
print("Original:", components)
print("Sorted:", sorted_components)
```

```python
#  Combine Multiple Electrical Values

# add voltage drops in series
voltages = [5, 12, 7]
total_voltage = (lambda values: sum(values))(voltages)
```

```python
print(" Combined Electrical Values ")
print("Individual Voltages:", voltages)
print("Total Voltage:", total_voltage)
```
