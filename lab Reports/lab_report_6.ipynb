Lab Report Tasks:  
Scenario:  
You are designing a small student management system. You need to maintain student data for a class. Each student has a name, a roll number, and a list of grades. The system must allow grade updates but keep roll numbers fixed.  
Tasks:  
1.Create a list of student names: ["Ali", "Usman", "Hassan", "Kamran"].  
2.Create a tuple of student roll numbers: (101, 102, 103, 104).  
3.For each student, create a list of grades (3 subjects) using nested lists:  
Example: [ [85, 90, 78], [88, 92, 80], ... ]  
4.Update Ali’s second grade to 95.  
5.Add a new grade for Hassan in the fourth subject.  
6.Attempt to change Kamran’s roll number and observe what happens. Explain why.  
7.Print all students' names with their updated grades in a readable format.  

```python
#  List of student names
student_names = ["Ali", "Usman", "Hassan", "Kamran"]
```

```python

#  Tuple of roll numbers 
student_rolls = (101, 102, 103, 104)
```

```python
#  Nested list of grades 
student_grades = [
    [85, 90, 78],   # Ali
    [88, 92, 80],   # Usman
    [75, 85, 89],   # Hassan
    [90, 87, 91]    # Kamran
]
```

```python
# Update Ali's second grade to 95
# -----------------------------
# Ali is at index 0, second grade is index 1
student_grades[0][1] = 95
```

```python
#  Add a new grade for Hassan (fourth subject)
# Hassan is at index 2
student_grades[2].append(93)
```

```python
# change Kamran's roll number
try:
    student_rolls[3] = 110
except TypeError as e:
    print("Error:", e)
    print("Explanation: Tuples are immutable, so roll numbers cannot be changed.")
```

```python
# Print all students' names with their updated grades

print("Student Grades ")
for i, name in enumerate(student_names):
    grades = student_grades[i]
    roll = student_rolls[i]
    print(f"Name: {name}, Roll No: {roll}, Grades: {grades}")
```
