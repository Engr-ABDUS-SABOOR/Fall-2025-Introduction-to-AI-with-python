Task 1:  Basic list comprehension  
Create a new list of numbers from 0 to 9 using a list comprehension, not a loop.  
basic syntax [expression for item in iterable]  

```python
# Task 1: Basic List Comprehension
# Create a list of numbers from 0 to 9

numbers = [x for x in range(10)]
print("Task 1 - Numbers 0 to 9:", numbers)
```

Task 2: Filtering items with if  
Given:  
nums = [1, 2, 3, 4, 5, 6]  
Use a list comprehension with an if clause to keep only the even numbers.  
 [expr for item in iterable if condition]  

```python
# Task 2: Filtering items with if
# Keep only even numbers

nums = [1, 2, 3, 4, 5, 6]
even_nums = [x for x in nums if x % 2 == 0]
print("Task 2 - Even Numbers:", even_nums)
```

Task 3:  Conditional expression inside comprehension  
Given:  
nums = [1,2,3,4]  
Create a new list where even numbers become "even" and odd numbers become "odd".  
expr_if_true if condition else expr_if_false inside comprehension  

```python
# Task 3: Conditional expression inside comprehension
# Even → "even", Odd → "odd"

nums2 = [1, 2, 3, 4]
even_odd = ["even" if x % 2 == 0 else "odd" for x in nums2]
print("Task 3 - Even/Odd Labels:", even_odd)
```

Task 4: Simple nested list comprehension  
Given a list of lists:  
data = [[1,2], [3,4]]  
Flatten it into a single list using a nested list comprehension.  

```python
# Task 4: Nested list comprehension (Flatten list)
# Flatten [[1,2], [3,4]] → [1,2,3,4]


data = [[1, 2], [3, 4]]
flattened = [item for sublist in data for item in sublist]
print("Task 4 - Flattened List:", flattened)
```
