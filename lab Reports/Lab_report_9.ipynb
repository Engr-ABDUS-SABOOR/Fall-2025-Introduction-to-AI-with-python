Current readings are stored in a list. Engineers want to calculate the square of each value for power loss calculation.  
Print the square of each current reading using formula I².  
Make a list of current values, loop and multiply each by itself.  

```python

# List of current readings in amperes
current_readings = [2, 3, 4, 5]  
```

```python
# Empty list to store squared values
current_squared = []
```

```python
# Loop through each current value and calculate I^2
for current in current_readings:
    current_squared.append(current ** 2)
```

```python
# Print the results
print("Current Squared Readings ")
print("Original Current Readings (A):", current_readings)
print("Squared Current Values (A^2):", current_squared)
```
