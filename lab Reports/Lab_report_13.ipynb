# Task 1: 
Create and Write Energy Data to a CSV File  
Create a file named energy_data.csv.  
Write the header and daily energy consumption records into the file using the CSV format:  
	Date  
	Energy consumed in kWh  
Each record must be written on a new line following the header  


```python
import csv

with open(r"C:\Users\PMLS\Desktop\ABDUS_SABOOR\AI_Lab\energy data.csv", mode="r") as file:
    reader = csv.reader(file)

    header = next(reader)   # read header
    print("Header:", header)

    for row in reader:
        print("Date:", row[0], "| Energy Consumed (kWh):", row[1])
```

# Task 2: 
Read and Display CSV File Data  
Read the contents of energy_data.csv and display all energy consumption records in a clear and readable format.  
The data must be read from the file, not hardcoded.  


```python
import csv

file_path = (r"C:\Users\PMLS\Desktop\ABDUS_SABOOR\AI_Lab\energy data.csv")

with open(file_path, mode="r") as file:
    reader = csv.reader(file)

    header = next(reader)   # read header
    print(f"{header[0]:<15}  {header[1]}")
    print("-" * 35)

    for row in reader:
        date = row[0]
        energy = row[1]
        print(f"{date:<15}  {energy} kWh")
```

# Task 3: 
Append New Energy Records  
Add new daily energy consumption records to energy_data.csv without removing existing data.  
Ensure:  
	The CSV format is preserved  
	New data is appended correctly  



```python
import csv

file_path = (r"C:\Users\PMLS\Desktop\ABDUS_SABOOR\AI_Lab\energy data.csv")

new_records = [
    ["2025-01-05", 15.3],
    ["2025-01-06", 13.7]
]

with open(file_path, mode="a", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(new_records)

print("New energy records appended successfully.")
```
