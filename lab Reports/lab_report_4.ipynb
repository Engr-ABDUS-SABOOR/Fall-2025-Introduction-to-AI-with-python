In circuit design, you often calculate series resistance.  
Write a function series() that takes a list or tuple of resistances.  
Return the total resistance.  
Add docstrings and annotations.  
Print the function’s __annotations__.  

```python
from typing import List, Tuple, Union

def series(resistances: Union[List[float], Tuple[float, ...]]) -> float:
    """
    Parameters:
    resistances (list or tuple of float): A list or tuple containing resistance values in ohms.

    Returns:
    float: Total series resistance in ohms.

    """
    total_resistance = sum(resistances)
    return total_resistance


```

```python
# Example usage
resistances_list = [5, 10, 15, 20]  
total = series(resistances_list)
```

```python
print("Resistances:", resistances_list)
print("Total Series Resistance:", total, "Ω")
```

```python
#  function annotations
print("Function Annotations:", series.__annotations__)
```
