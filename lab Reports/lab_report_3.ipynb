A resistor with zero value creates a short circuit, which can damage the circuit. Students need to test the resistance before use.  
Check if the resistance value is zero. If yes, print "Short Circuit", else print "Safe".  
Store resistance in a variable, use if-else.  

```python
# Store the resistance value (in ohms)
resistance = 0  

print(" Resistance Short Circuit Check ")
# Check for short circuit
if resistance == 0:
    print("Short Circuit")
else:
    print("Safe")
```
