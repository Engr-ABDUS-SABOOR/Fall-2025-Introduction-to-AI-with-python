Lab Report Task:  
You are developing a Student Result Processing System for a small institute.  
The program takes student marks as input and calculates results. Since users may enter incorrect data, the program must handle errors properly using Python exception handling.  


```python
"""
Student Result Processing System

This program calculates a student's total marks and percentage,
determines pass/fail, and handles invalid inputs using exception handling.
"""

def get_mark(subject):
    """
    Get valid mark for a subject.
    Raises ValueError if mark is not between 0 and 100.
    """
    while True:
        try:
            mark = float(input(f"Enter marks for {subject} (0-100): "))
            if mark < 0 or mark > 100:
                raise ValueError(f"Invalid mark for {subject}. Must be 0-100.")
            return mark
        except ValueError as error:
            print("Error:", error)
        except TypeError:
            print("Error: Please enter a numeric value.")

def main():
    print("=== Student Result Processing System ===")
    student_name = input("Enter student name: ")

    # List of subjects
    subjects = ["Math", "Science", "English"]
    marks = []

    # Step 1 & 2: Input marks with exception handling
    for subject in subjects:
        marks.append(get_mark(subject))

    # Step 3: Calculate total and percentage
    total_marks = sum(marks)
    max_total = 100 * len(subjects)
    percentage = (total_marks / max_total) * 100

    # Step 4: Determine result
    result = "Pass" if percentage >= 50 else "Fail"

    # Step 5: Display result
    print("\n=== Student Result ===")
    print("Name:", student_name)
    print("Total Marks:", total_marks, "/", max_total)
    print("Percentage:", f"{percentage:.2f}%")
    print("Result:", result)

if __name__ == "__main__":
    main()
```
