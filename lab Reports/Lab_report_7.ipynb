# **Lab_Report_7**:  
You are working in a university electronics lab.   
You are asked to monitor a small solar home system installed on the lab roof. The system has:  
a)A solar panel  
b)A battery  
c)A load controller  
d)A temperature sensor  
Each minute, the system sends a small dictionary of readings  
I.Create a dictionary named reading1 containing the 5 fields shown in the scenario.  
II.From reading1, print:  
i.The solar power  
ii.The battery voltage  
iii.The load status  
III.Print how many fields the dictionary contains after your modifications.  
IV.Remove the "panel_current" field.  
V.Get Keys, Values, and Items  
Print:  
i.All keys  
ii.All values  
iii.All key-value pairs  
iv.Use .keys(), .values(), and .items().  


```python
# Task I: Create the dictionary "reading1"


reading1 = {
    "solar_power": 320,        # Watts
    "battery_voltage": 12.5,   # Volts
    "load_status": "ON",       # Load controller state
    "temperature": 34.2,       # Celsius
    "panel_current": 8.4       # Amps
}
```

```python
# Task II: Print required fields

print("Solar Power:", reading1["solar_power"])
print("Battery Voltage:", reading1["battery_voltage"])
print("Load Status:", reading1["load_status"])
```

```python
# Task III: Print number of fields

print("Total Fields:", len(reading1))
```

```python
# Task IV: Remove 'panel_current' field

reading1.pop("panel_current")

print("After Removing 'panel_current':", reading1)
```

```python
# Task V: Print Keys, Values, and Items

print("All Keys:", reading1.keys())
print("All Values:", reading1.values())
print("All Key-Value Pairs:", reading1.items())
```
