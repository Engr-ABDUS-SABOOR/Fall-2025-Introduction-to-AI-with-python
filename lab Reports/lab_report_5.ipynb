Lab Report Tasks:  
 sum: sum() adds up numbers in a list.  
i.The power consumed by individual components is listed in a list. Find the total power consumed by adding readings.  
ii.A list contains the values of a series of connected resistances. Find its equivalent resistance using the sum.  
Example:  
nums = [1, 2, 3]  
print(sum(nums))  
Output:  
6  

```python
# List of power consumed by individual components in watts

power_readings = [10, 15, 20, 25]  
```

```python
# Calculate total power using sum()

total_power = sum(power_readings)
```

```python
print("Total Power Consumption")
print("Power readings (W):", power_readings)
print("Total Power Consumed (W):", total_power)
```

```python
#  Equivalent Series Resistance

# List of resistances in ohms
resistances = [5, 10, 15, 20]  
```

```python
# Calculate total (equivalent) series resistance using sum()
total_resistance = sum(resistances)
```

```python
print("Equivalent Series Resistance ")
print("Resistances (立):", resistances)
print("Total Series Resistance (立):", total_resistance)
```
